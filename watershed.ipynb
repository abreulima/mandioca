```python
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("Araruta.png")
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray, cmap=plt.cm.gray)
```

```python
ret, thresh2 = cv.threshold(gray, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
ret, thresh3 = cv.threshold(gray, 0, 255, cv.THRESH_BINARY + cv.THRESH_TRIANGLE)
```

```python
plt.figure("OTSU")
plt.imshow(thresh2, cmap=plt.cm.gray)

plt.figure("Triangle")
plt.imshow(thresh3, cmap=plt.cm.gray)

plt.show()
```
